```python
VERSION = '18_v1_05'
FOLD_PATH = '../../00_EDA/00_v2_07/result/'
MODEL = 'microsoft/deberta-xlarge'
LR = 6e-6
HEAD_LR = 6e-6
SEED = 100
TRN_BS = 1
VAL_BS = 1
ACCUM_STEP = 1
EPOCHS = 5
STOP_EPOCH = 5
RESTART = 1
HIDDEN_DROP_PROB = 0
P_DROP = 0
RNN = 'none'
WARMUP_RATIO = 0.1
LOSS = 'xentropy'
HEAD = 'simple'
AUG = 'false'
MIXUP_ALPHA = 1.0
P_AUG = 0
AUG_STOP_EPOCH = 2
MSD = 'true'
MULTI_LAYERS = 1
EVAL_STEP = -1
NUM_LABELS = 3
NUM_LABELS_2 = 7
ADV_SIFT = 'false'
FP16 = 'true'
WD = 0.01
FREEZE = 'true'
MULTI_TASK = 'false'
W_MT = 1.0
AWP = 'true'
AWP_LR = 5e-4
AWP_EPS = 1e-3
AWP_START_EPOCH = 1
PRETRAINED_DETECTOR_PATH = f'../../input/tascj/result/deberta_xlarge_fold0.pth'
MASK_PROB = 0.8
MASK_RATIO = 0.3
SCHEDULER = 'cosine_hard'
NUM_CYCLES = EPOCHS
CP = 'true'
WINDOW_SIZE = 1024
INNER_LEN = 768
EDGE_LEN = 128
BNB = 32
```
